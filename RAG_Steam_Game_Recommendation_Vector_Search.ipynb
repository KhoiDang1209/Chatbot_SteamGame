```python
import pandas as pd
data=pd.DataFrame(list(collection.find()))
data.head()
```

```python
data.drop('_id', axis=1, inplace=True)
data.head()
```

```python
from sentence_transformers import SentenceTransformer, util
embedding_model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
def get_embedding(text) -> list[float]:
    if not isinstance(text, str) or not text.strip():
        print(f"Skipping invalid text: {text}")  # Debugging message
        return []
    return embedding_model.encode(text).tolist()
```

```python
data['embedding_description']=data['description'].apply(get_embedding)
data['embedding_tags']=data['tags'].apply(get_embedding)
data['embedding_poppular_tags']=data['popular_tags'].apply(get_embedding)
```
