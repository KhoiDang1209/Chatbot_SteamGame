```python
import pandas as pd
data=pd.DataFrame(list(collection.find()))
data.head()
```

```python
data.drop('_id', axis=1, inplace=True)
data.head()
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Step 1: Combine game fields into a single text block
def combine_game_fields(game):
    fields = [
        str(game.get("name", "")),
        str(game.get("description", "")),
        str(game.get("all_reviews", "")),
        str(game.get("release_date", "")),
        str(game.get("developer", "")),
        str(game.get("publisher", "")),
        str(game.get("price", ""))
    ]
    return " ".join(fields)

# Step 2: Split the combined text into overlapping chunks
def split_game_text(game, chunk_size=400, overlap=50):
    full_text = combine_game_fields(game)
    splitter = RecursiveCharacterTextSplitter(
        chunk_size=chunk_size,
        chunk_overlap=overlap,
        separators=["\n\n", "\n", ".", " ", ""]
    )
    return splitter.split_text(full_text)
data['text_chunks'] = data.apply(split_game_text, axis=1)
```

```python
data.head()
```

```python
from sentence_transformers import SentenceTransformer, util
embedding_model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
def get_embedding(text) -> list[float]:
    if not isinstance(text, str) or not text.strip():
        print(f"Skipping invalid text: {text}")  # Debugging message
        return []
    return embedding_model.encode(text).tolist()
```

```python
data['embedding_chunks']=data['text_chunks'].apply(lambda x: [get_embedding(text) for text in x])
```
